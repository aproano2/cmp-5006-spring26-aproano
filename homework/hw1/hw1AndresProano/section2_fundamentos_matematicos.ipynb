# 2. Algoritmo de Euclides

## 2.1 Calcular: $\gcd(1160, 480)$

Aplicamos el Algoritmo de Euclides:

- $1160 = 480(2) + 200$
- $480 = 200(2) + 80$
- $200 = 80(2) + 40$
- $80 = 40(2) + 0$

El último residuo distinto de cero es 40, por tanto:

$$
\boxed{\gcd(1160, 480) = 40}
$$


```python
# Verificación 2.1
import math
math.gcd(1160, 480)
```

## 2.2 Algoritmo de Euclides Extendido

Buscamos $x$ e $y$ tales que:

$$
1160x + 480y = \gcd(1160, 480) = 40
$$

A partir de las divisiones:

1) $1160 = 480(2) + 200$  
2) $480 = 200(2) + 80$  
3) $200 = 80(2) + 40$

Retro–sustitución:

De (3):

$$
40 = 200 - 80(2)
$$

De (2): $80 = 480 - 200(2)$

Sustituyendo:

$$
40 = 200 - (480 - 200(2))(2) = 5(200) - 2(480)
$$

De (1): $200 = 1160 - 480(2)$

Sustituyendo:

$$
40 = 5(1160 - 480(2)) - 2(480) = 5(1160) - 12(480)
$$

**Resultado:**

$$
\boxed{x = 5,\quad y = -12}
$$

Verificación:

$$
1160(5) + 480(-12) = 40
$$


```python
# Verificación 2.2 (extendido)
def egcd(a, b):
    """Devuelve (g, x, y) tal que ax + by = g = gcd(a,b)."""
    if b == 0:
        return (a, 1, 0)
    g, x1, y1 = egcd(b, a % b)
    x = y1
    y = x1 - (a // b) * y1
    return (g, x, y)

g, x, y = egcd(1160, 480)
g, x, y, 1160*x + 480*y
```

Factorización:

$$

$$

$$

$$

$$

**Resultado:**

$$
\boxed{\varphi(360)=96}

# 4. Teoría de la Información – Entropía de Shannon

La entropía de Shannon se define como:

$$
H(X) = -\sum_x p(x)\log_2 p(x)
$$

Dado:

- $P(A)=\frac{1}{2}$
- $P(B)=\frac{1}{4}$
- $P(C)=\frac{1}{8}$
- $P(D)=\frac{1}{8}$

Calculamos término por término:

- Para $A$: $-\frac{1}{2}\log_2(\frac{1}{2}) = \frac{1}{2}$
- Para $B$: $-\frac{1}{4}\log_2(\frac{1}{4}) = \frac{1}{2}$
- Para $C$: $-\frac{1}{8}\log_2(\frac{1}{8}) = \frac{3}{8}$
- Para $D$: igual que $C$ $\rightarrow \frac{3}{8}$

Sumando:

$$
H(X)=\frac{1}{2}+\frac{1}{2}+\frac{3}{8}+\frac{3}{8}=1.75\ \text{bits}
$$

**Resultado:**

$$
\boxed{H(X)=1.75\ \text{bits}}
$$


```python
# Verificación 4 (cálculo directo)
import math
p = {"A": 1/2, "B": 1/4, "C": 1/8, "D": 1/8}
H = -sum(pi * math.log(pi, 2) for pi in p.values())
H
```

# 5. Teorema de Bayes – IDS

Queremos:

$$
P(A\mid F)
$$

Donde:
- $A$: ocurre un ataque real
- $F$: el IDS levanta una alerta (flag)

Datos:
- $P(A)=0.001$
- $P(F\mid A)=0.99$ (verdadero positivo)
- $P(F\mid \neg A)=0.01$ (falso positivo)
- $P(\neg A)=1-P(A)=0.999$

Aplicamos Bayes:

$$
P(A\mid F)=\frac{P(F\mid A)P(A)}{P(F\mid A)P(A)+P(F\mid \neg A)P(\neg A)}
$$

Sustituyendo:

$$
P(A\mid F)=\frac{0.99(0.001)}{0.99(0.001)+0.01(0.999)}
$$

Cálculo:
- Numerador: $0.99\times 0.001=0.00099$
- Denominador: $0.00099+0.01\times 0.999 = 0.01098$

Entonces:

$$
P(A\mid F)=\frac{0.00099}{0.01098}\approx 0.0901 \approx 9\%
$$

**Resultado:**

$$
\boxed{P(A\mid F)\approx 9\%}
$$

**Nota de seguridad:** Este resultado ilustra la *falacia de la tasa base*: aunque el IDS sea “muy bueno”, cuando los ataques son raros, una gran fracción de las alertas siguen siendo falsos positivos.
